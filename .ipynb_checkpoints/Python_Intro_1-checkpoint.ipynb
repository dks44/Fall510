```python
import urllib.request #this request will retrieve the Course Schedule file from Github and save a copy in a pre-determined location under a specified file name
r = urllib.request.urlopen('https://github.com/Math510/Math510.github.io/raw/master/2015%20FALL%20510%20Course%20Schedule.pdf')
# SW: Open the course schedule on GitHub and assign the results to the variable r
myfile = open('C:\\Users\\Deeza\\Desktop\\Math 510 Syllabus.pdf', 'wb') #we will save the doc from Github to my desktop with a different name
# SW: Specify the 'w' option so the file is writable
# SW: Specify the 'b' option because the file type is PDF (not purely text)
myfile.write(r.read()) #writing the original doc from Github and reading to my desktop
# SW: not quite, this reads the file stored in the variable r into memory and then writes it to myfile
myfile.close() # SW: Close myfile so it's unlocked and available for other manipulation or other applications.
```

## Modules

### Create a variable

```python
title = 'Fairy' #created new variable called Fairy
# SW: this creates a variable named title and assigns to it the string 'Fairy' (-1)
print(title) #output the variable name
```

### Import variables from a module

```python
import Class_1_Module #copied Class_1_Module from Python folder into Notebook, then ran code.
# SW: this loads the module
dir(Class_1_Module)
print(Class_1_Module.title) #outputs the title defined in the Class_1_Module
print(Class_1_Module.MyAge) #outputs the results of the calculation in the Class_1_Module

print(title) #still prints title as defined above since this is the 'title' variable recognized
```

## Show the difference in scope between importing a module and importing a specific variable from the module

```python
#copies into local name space, overwrites same named variable
#without warning
from Class_1_Module import title #imported title from Class_1_Module and now overwrites original 'title' variable created previously
print(title) #prints title from Class_1_Module
```
